```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/cleanclean_cornell.csv')
```

```python
df['ipm_county'] = df['ipm_county'].replace('Baltimore', 'Baltimore City')
print(df[df['ipm_county'] == 'Baltimore'])
```

```python
df['ipm_county'] = df['ipm_county'].replace{
    'Caroll County','Carroll County'
    'City of Waynesboro','Waynesboro City'
    'City of Staunton','Staunton'
    'City of Manassas Park','Manassas Park'
    'City of Manassas',''
    'City of Lynchburg',''
    'City of Lexington',''
    'City of Harrisonburg',''
    'City of Buena Vista',''
}
```

```python
df['ipm_county'] = df['ipm_county'].replace({
    'City of Waynesboro':'Waynesboro City',
    'City of Staunton':'Staunton City',
    'City of Manassas Park':'Manassas Park City',
    'City of Manassas':'Manassas City',
    'City of Lynchburg':'Lynchburg City',
    'City of Lexington':'Lexington City',
    'City of Harrisonburg':'Harrisonburg City',
    'City of Buena Vista':'Buena Vista City'
})
df.loc[
    (df['ipm_state'] == 'Kentucky') & (df['ipm_county'] == 'Caroll County'),
    'ipm_county'
] = 'Carroll County'
```

```python
print(df[(df['ipm_state'] == 'Kentucky') & (df['ipm_county'].str.contains('Carroll'))])
```

```python
df = df.drop()
```
