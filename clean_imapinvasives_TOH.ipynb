```python
#looking at TOH data and perhaps merging two suvey sets cuz why the hell not
import pandas as pd
import matplotlib as plt
import numpy as np
import seaborn as sn
```

```python
df = pd.read_csv('/content/PRESENCE_LINE_CONFIRMED__2025_09_12_141635529.csv', encoding="cp1252")
```

```python
#getting rid of any plant that isnt the TOH
df = df[df["scientific_name"] == "Ailanthus altissima"]
```

```python
#getting rid of the same tree id just in case
df = df.drop_duplicates(subset="present_species_id", keep="first")
df
```

```python
#getting just the columns im interested in
df = df[['centroid_x','centroid_y','scientific_name','common_name','observation_date','number_found']]
```

```python
#changing observation date into something I like better
df["observation_date"] = pd.to_datetime(df["observation_date"], format="%d-%b-%Y")
```

```python
#gonna extract y m d bc thats how my other data is
df["Year"] = df["observation_date"].dt.year
df["Month"] = df["observation_date"].dt.month
df["Day"] = df["observation_date"].dt.day
```

```python
#dont need these anymore
df= df.drop(['observation_date','Day'], axis=1)
#lol wont let me run thru this
```

```python
#dont know what NaN in number_found means so we're losing those too
df = df.dropna(subset=['number_found'])
```

```python
#saving
df.to_csv("clean_imap_TOH.csv", index=False, encoding="cp1252")
```
