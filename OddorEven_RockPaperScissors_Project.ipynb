```python
#Odd or Even?
# Ask the user for a number. Depending on whether the number is even or odd,
# print out an appropriate message to the user

#to start
prompt = "Type 'start' to play Odds or Evens: "
message = ''

while message != 'start':
    message = input(prompt)
    print(message)

#to learn
print('Please enter a number: ')
number = int(input())
if (number %2 == 0):
  print('Your number is even.'),
if (number %2 == 1):
    print('Your number is odd.')
```

```python
#Rock, Paper, Scissors- Shoot!

#To start
prompt = "Type 'start' to play: "
message = ''

while message != 'start':
    message = input(prompt)
    print(message)

#Pleasentries
print('Welcome! You are playing Mermaid')

#Rule Book
win_rules = {
    'rock': 'scissors',
    'scissors' : 'paper',
    'paper': 'rock'
}

#while (some condition):  # TODO: figure out how to complete this while loop to make it run indefinitely
    #To play
    playrps = "Select your object from rock, paper, or scissors. Now, shoot!: "
    print(playrps)
    message = ''

    #Until Valid
    while message not in ('rock', 'paper', 'scissors'):
        message = input()
        # print("Mermaid wins this time!")

    #How Mermaid plays RPS
    import random
    mermaid = random.choice(list(win_rules))
    print(f"Mermaid played {mermaid}")

    #Finding a Winner
    if win_rules [mermaid] == message:
        print('Mermaid wins!')
    elif win_rules [message] == mermaid:
        print('You win!')
    else:
        print('It is a tie!')

playagain = 'Mermaid wins!', 'You win!', 'It is a tie!'
while playagain in ('Mermaid wins!', 'You win!', 'It is a tie!'):
    print('Again?')



# Endless or Not
# #play_again = input('Do you want to play again?')
# if play_again != 'yes':
#     print('Goodbye!')
#     break
```
