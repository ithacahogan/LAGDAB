```python
#ok now we see what this TOH dataset looks like
import pandas as pd
import matplotlib as plt
import numpy as np
import seaborn as sn
```

```python
#importing csv, need to reassign the first row as the column names using header
df = pd.read_csv('/content/distribution_TOH_notclean - Sheet1.csv', header=0)
```

```python
#this is a lot different from the imap invasives data.
#imap shows exactly where it is like x,y coordinates
#need to assign NaN and pull out data

df['County'] = df['County'].replace("", np.nan)
```

```python
#dropping nan values
df = df.dropna(subset=['County'])

#dropping entire columns
df= df.drop(['Symbol'], axis=1)
```
