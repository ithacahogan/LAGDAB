```python
#
import pandas as pd
import matplotlib as plt
import numpy as np
import seaborn as sn
```

```python
df = pd.read_csv('/content/lyde.csv')
```

```python
#dropping na values
df = df.dropna(subset=['lyde_density'])
```

```python
#dropping entire columns
df = df= df.drop(['source','survey','year','source_agency','pointID'], axis=1)
```

```python
df
```

```python
#gotta make sure that present & established are boolean dtypes
df[['lyde_present','lyde_established']] = df[['lyde_present','lyde_established']].astype(bool)
```

```python
#changing a column name for clarity's sake
df.rename(columns={"bio_year": "lyde_year"}, inplace=True)
```

```python
#ok that seems clean and whatnot...
#saving
df.to_csv("clean_lyde.csv", index=False)

#how to find the file againnnnnn
from google.colab import files
files.download("clean_lyde.csv")
```
