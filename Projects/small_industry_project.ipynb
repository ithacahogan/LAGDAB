```python
#well as i looked earlier I think all of the numeric columns are already ints and floats but lets do it just in case
#to_numeric
numberstuff = ['release_lbs','latitude','longitude','year_x','year_y']
for col in numberstuff:
  df[col] = pd.to_numeric(df[col], errors = "coerce")

#fill the na values
numberstuff = ['release_lbs','latitude','longitude','year_x','year_y']
df[numberstuff] = df[numberstuff].fillna(0)

#making sure its the right kind of data type
makeitfloat = ['release_lbs','latitude','longitude']
df[makeitfloat] = df[makeitfloat].astype(float)

df['year_x'] = df['year_x'].astype(int)
df['year_y'] = df['year_y'].astype(int)
```

```python
#part 2: Exploratory Data Analysis (EDA)

#Q1: What is the total amount of chloroprene released per year in St. John the Baptist Parish?
#A1: The outputs below show the total amount of chloroprene that is released, per year, in SJBP
#chloroprene_in_sjbp = df2.groupby([])
sjbp = df2['county'] == 'St. John the Baptist'
chemc = df2['chemical'] == 'Chloroprene'
df2_parsed = df2[sjbp & chemc]

sj_chemc = df2_parsed.groupby(['county','chemical','year'])['release_lbs'].sum().sort_values(ascending=False).reset_index()

print(sj_chemc)
```

```python
#Q2: Which facility has released the most chloroprene (cumulative)?
#A2: The Gulf Elastomers Plant, facility F00098, has released a cumulative total of 257,581 lbs of chloroprene from 2012-2024.

chemc = df2['chemical'] == 'Chloroprene'
df2_parsed = df2[chemc]

facil_chemc = df2_parsed.groupby(['facility_id','facility_name'])['release_lbs'].sum().sort_values(ascending=False).reset_index()
print(facil_chemc)
```

```python
#Q3: How do chloroprene releases compare to other chemicals in this dataset?
#A3: Chloroprene is the leading chemical by pounds that are released, making it a major concern. In the 12 year
## span we are evaluating, 98 million pounds of chloroprene have been released, compared to the second most released chemical
### Toluene, which had 12 million pounds released in the same span.

all_chem = df2.groupby('chemical')['release_lbs'].sum().sort_values(ascending=False).reset_index()
print(all_chem)
```

```python
#Q4: What is the average release per facility in the parish?
#A4: The average release per facility in SJtBP is 61,309 lbs of chemicals over 12 years.

#this shows the average
sjbp = df2['county'] == 'St. John the Baptist'
sjbp_AR = df2[sjbp]
avg_release = sjbp_AR.groupby(['facility_id','facility_name'])['release_lbs'].mean().sort_values(ascending=False).reset_index()
print(avg_release)

#Q4 Continued

avg_release2 = sjbp_AR.groupby('facility_id')['release_lbs'].mean().reset_index()
avg_release_overall = avg_release2['release_lbs'].mean()
print(avg_release_overall)
```

```python
#Q5: Which year saw the highest overall chemical release?
#A5: 2012 had the highest overall chemical release, with 13.8 million chemical pounds released.

year_high = df2.groupby('year')['release_lbs'].sum().sort_values(ascending=False).reset_index()
print(year_high)
```

```python
#Part 3 (save each as a png file)

#Q1: Bar chart: Total chloroprene releases by year.
chemc = df2['chemical'] == 'Chloroprene'
df2_chlor = df2[chemc]

chlor_byyear = df2_chlor.groupby('year')['release_lbs'].sum().div(1000000).reset_index()

plt.bar(chlor_byyear['year'],chlor_byyear['release_lbs'])
plt.xlabel('Year')
plt.ylabel('Chloroprene Release (millions of lbs)')
plt.title('Chloroprene Released from 2012-2024')
plt.show()
```

```python
#Q2: Pie chart: Share of chloroprene vs other chemicals released.

df2_nochlo = df2[df2['chemical'] != 'Chloroprene']

chl_v_chems = df2.groupby('chemical')['release_lbs'].sum()
chl_v_chems
```

```python
df2_sochlo = df2[df2['chemical'] == 'Chloroprene']

wee = df2_nochlo['release_lbs'].sum()
chlochlo = df2_sochlo['release_lbs'].sum()

chems = [wee, chlochlo]
stuff = ['Other Chemicals', 'Chloroprene']
plt.pie(chems,labels=stuff,autopct='%1.1f%%'
    , colors=['olive','tab:blue'])
plt.title('Share of Chemicals Released from 2012-2024')
plt.show()
```

```python
#Q3: Line chart: Trend of chloroprene releases over time (multi-year).

plt.plot(chlor_byyear['year'],chlor_byyear['release_lbs'])
plt.xlabel('Year')
plt.ylabel('Chloroprene in millions of lbs')
plt.title('Chloroprene Released from 2012-2024')
plt.show()
```
