```python
#nested dictionaries
theater = {
    'candy': 'sourpatch',
    'popcorn' :'ranch',
    'soda' : 'coke'
}

movies = {
    'horror' : 'attack_monkey',
    'comedy':'skipping_lanes',
    'indie':'sunflower_hair'
}

weather = {
    'outside':'rainy',
    'temperature':'52F',
    'timeofday':'dusk'
}

nested = {
    'snacks': theater,
    'movie': movies,
    'because': weather
}

for key, value in nested.items():
  print('Here is your night: '+ key, nested[key])
```
